### 1. NumPy
***

NumPy (**Numerical Python**) es la columna vertebral de todo el ecosistema de datos en Python.

Problema:

```python
a = [1, 2, 3]
b = [4, 5, 6]

resultado = []
for i in range(len(a)):
    resultado.append(a[i] + b[i])
```

¿Se puede hacer esto más rápido y con menos código?

Las listas de Python son potentes porque pueden guardar cualquier tipo de dato (objetos), pero eso las hace lentas, debido a que cada elemento en una lista de Python es un objeto completo con metadatos.

```python
[1, 4, "Hola", 9, "mundo", [2, 5], True]
```

NumPy introduce el *n-dimensional* array, donde todos los elementos deben ser del mismo tipo () y se almacenan físicamente uno al lado del otro en la RAM, permitiendo que el procesador acceda a ellos a una mayor velocidad.

#### 1.3. Vectorización

NumPy en lugar de procesar elementos uno por uno con bucles, delega la operación a bibliotecas de bajo nivel

##### Ejemplo 1

```python
precios = [100, 200, 300, 400]
con_impuesto = []
for p in precios:
    con_impuesto.append(p * 1.10)

print(con_impuesto)
```

```python
precios = np.array([100, 200, 300, 400])
con_impuesto = precios * 1.10
print(con_impuesto)
```

```python
precios2 = [100, 200, 300, 400]
```

```python
precios2 * 2
```

##### Ejemplo 2

```python
angulos = np.array([0, np.pi/2, np.pi])
resultados = np.sin(angulos)
print(resultados)
```

##### Ejemplo 3: Producto Punto

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
resultado = np.dot(a, b)
print(resultado)
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
dot = 0
for i in range(len(a)):
    dot += a[i] * b[i]

print(dot)
```

#### 1.4. Operaciones

##### Ejemplo 4:

```python
# Comparaciones lógicas
notas = np.array([4.5, 7.8, 9.2, 5.0, 3.2])
resultados = notas >= 7.0
print(resultados)
```

```python
# Operaciones estadísticas

resultados = np.mean(notas)
print(resultados)
```

```python
resultados = np.median(notas)
print(resultados)
```

```python
resultados = np.std(notas)
print(resultados)
```

```python
resultados = np.sum(notas)
print(resultados)
```

```python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print(matriz)
```

```python
escalar = 10
print(matriz + escalar)
```

```python
# Operaciones por ejes
print(np.sum(matriz, axis=0)) # Suma las columnas
print(np.sum(matriz, axis=1)) # Suma las filas
```

#### 1.4. Ejemplos

```python
presion = np.array([101.3, 102.5, 900.0, 101.8, 288.5, 103.0])

# 1. Crear una lista para valores menores a 200
mascara = presion < 200
print(mascara)
```

```python
# 2. Aplicar
datos_limpios = presion[mascara]
print(datos_limpios)
```

```python
# 3. Reemplazar mayores a 200 por el promedio de los sanos
promedio = np.mean(datos_limpios)
presion[presion > 200] = promedio

print(f"Datos corregidos: {presion}")
```
