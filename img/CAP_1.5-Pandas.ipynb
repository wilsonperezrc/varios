#### Pandas Series
Es muy parecido a un array de una dimensión (o vector) de NumPy.

- Arreglo unidimensional indexado
- Búsqueda por índice
- Operaciones aritméticas
- Distintos tipos de datos

```python
list = pd.Series(['Pedro','Joana','Angel','Paco'], index=[1,7,10,30])
# Aquí definimos una lista con índices específicos
```

#### Pandas DataFrame

Parecido a las estructuras matriciales en NumPy.

- Arreglo de dos dimensiones
- Búsqueda por índice (columnas o filas)
- Distintos tipos de datos
- Tamaño variable


```python
# Se crea un diccionario
dict = {'Jugador':['Alex','Katy','Ana','Jorch'],
 'Altura':[182.0,156.23,162.0,188.0],
 'Puntos':[100,200,50,300]}

dict
```

```python
# Se crea un dataframe
df = pd.DataFrame(dict)

df
```

```python
df2 = pd.DataFrame(dict, index=[2,3,5,7])

df2
```

```python
df.columns
```

```python
df.index
```

```python
df2.index
```

#### Filtrado con loc y iloc

Navegar por un dataFrame por funciones que permiten filtrar datos de manera más específica

 - .loc
 - .iloc


#### .loc: Filtra según un label



```python
df.loc[:]
# Para poder visualizar todo
```

```python
df.loc[0:4]
# Filtrar por medio de indices.
```

```python
df.loc[0:4, ['Titulo', 'Autor']]
# Filtrar a traves de etiquetas. Y de manera indexada.
```

```python
# Operación con los valores.
df.loc[:,['Precio']] * 2
```

```python
# Filtrar con operaciones lógicas.
df.loc[:,['Autor']] == 'JJ Smith'
```

#### .iloc: Filtrar por índices

```python
# Obtener todos los datos
df.iloc[:]
```

```python
df.iloc[:,1:3]
```

```python
df.iloc[:2,]
```

```python
df.iloc[:2,[1,2]]
```

```python
df.iloc[[1],[1]]
```

```python
df.iloc[1,1]
```

```python
df.iloc[1]
```

#### Manejo de datos nulos

```python
dict = {
    'Col1':[1,2,3,np.nan],
    'Col2':[4,np.nan,6,7],
    'Col3':['a','b','c',None]
}

df = pd.DataFrame(dict)
df
```

```python
# Retorna una lista con datos booleanos. True hay valor, False no lo hay.
df.isnull()
```

```python
df.isnull()*1
```
